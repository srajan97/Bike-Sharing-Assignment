# Feature Scaling

To make all the features in same scale for an easy interpretation.

```python
#Importing the required library:

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler
from sklearn.feature_selection import RFE
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
```

```python
#Now scaling continuous variables:
numeric_vars = ['temp','atemp','hum','windspeed','cnt']

scaler = MinMaxScaler()

#Fit and transform of training set only:
new_train[numeric_vars] = scaler.fit_transform(new_train[numeric_vars])
```

```python
#Inspect stats after scaling:

new_train.describe()
```

```python
new_train.head()
```

## For the model building,dividing into X and Y sets:

```python
y_train = new_train.pop('cnt')
X_train = new_train
```

### Recursive feature elimination

```python
#Importing RFE and LinearRegression

from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
#Running RFE with the output number of the variable equal to 15

da = LinearRegression()
da.fit(X_train, y_train)

rfe = RFE(da, 15)             # running RFE
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
col
```

```python
X_train.columns[~rfe.support_]
```

```python
#Creating X_test dataframe with RFE selected variables
X_train_rfe = X_train[col]
```

# Building Linear Model using 'STATS MODEL'


### Model 1
### VIF Check

```python
#Checking for the VIF values of the feature variables:

from statsmodels.stats.outliers_influence import variance_inflation_factor

#Now Creating a dataframe that will contain the names of all the feature variables and their respective VIFs:

vif = pd.DataFrame()
vif['Features'] = X_train_rfe.columns
vif['VIF'] = [variance_inflation_factor(X_train_rfe.values, i)
              for i in range(X_train_rfe.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
import statsmodels.api as sm

#Adding a constant:
X_train_lm1 = sm.add_constant(X_train_rfe)

#Creating a first fitted model:
df = sm.OLS(y_train, X_train_lm1).fit()
```

```python
#Checking the parameters we got:

df.params
```

```python
#Printing a summary of the linear regression model which we got:

print(df.summary())
```
